### Python se Excel Automation - Salary Sheet + Bonus + Total + Formatting

```python
# Excel file Create + Sheet Rename

from openpyxl import Workbook   # Use 'Workbook' with capital W

# Create a new workbook
wb = Workbook()

# Get the active sheet
ws = wb.active

# Rename the sheet
ws.title = "Salary"

# Save the workbook
wb.save("C:\\DATA ANALYST\\Projects\\Python\\Salary.xlsx")
```

```python
# Add header row
header = ['Emp_Id','Name','Basic_Salary','Bonus','Total Salary']
ws.append(header)

# Save to your chosen folder
wb.save("C:\\DATA ANALYST\\Projects\\Python\\Salary.xlsx")
```

```python
# Employee Data
data = [
    [101, 'Rohit Kumar', 25000],
    [102, 'Abhishek Singh', 34000],
    [103, 'Sunil Yadav', 67000],
    [104, 'Mandeep Singh', 56000]
]

# Append each row
for row in data:
    ws.append(row)

# Save the workbook
wb.save("C:\\DATA ANALYST\\Projects\\Python\\Salary.xlsx")
```

```python
# Bonus Calculation (12% of Basic Salary)

for i in range(2,ws.max_row+1):
    ws[f"D{i}"]= f"=C{i}*12/100"

# Explanation
# - range(2, ws.max_row+1) → starts from row 2 (since row 1 is usually the header) and goes until the last row.
# - ws[f"D{row}"] → writes into column D (Bonus column).
# - f"=C{row}*12/100" → inserts an Excel formula that calculates 12% of the value in column C (Basic_Salary).

wb.save("C:\\DATA ANALYST\\Projects\\Python\\Salary.xlsx")
```

```python
# Total Salary Calculation

for i in range(2,ws.max_row+1):
    ws[f"E{i}"]=f"=C{i}+D{i}"

# 🔎 Explanation
# - range(2, ws.max_row+1) → loops through all rows starting from row 2 (since row 1 is header).
# - ws[f"E{i}"] → writes into column E (Total Salary).
# - f"=C{i}+D{i}" → inserts an Excel formula that adds Basic Salary (C column) and Bonus (D column).


wb.save("C:\\DATA ANALYST\\Projects\\Python\\Salary.xlsx")
```

```python
# Header Formating (Bold + Color)

from openpyxl.styles import Font , Alignment,PatternFill

# Apply bold font and center alignment to header cells

for col in ws[1]:   # First Row = header
    col.font=Font(bold=True, color="FFFFFF")
    col.fill=PatternFill(start_color="0000FF", end_color="0000FF", fill_type="solid") 
    col.alignment= Alignment(horizontal="center")


# Save workbook
wb.save("C:/DATA ANALYST/Projects/Python/Salary.xlsx")
```

```python
# Auto fit Columns width

from openpyxl.utils import get_column_letter

for col in range(1,ws.max_column+1):
    column_letter=get_column_letter(col)
    ws.column_dimensions[column_letter].width=15


# Save workbook
wb.save("C:/DATA ANALYST/Projects/Python/Salary.xlsx")

```
